```python
fn = "data/01.txt"
```

```python
def max_consec( fn ):
    with open(fn, "r") as f:
        best = cur = 0
        while True:
            x = f.readline()
            # test for "" in case no blank line at end of input
            if x == "\n" or x == "":
                print(cur)
                if cur > best:
                    print("*** NEW BEST ***")
                    best = cur
                cur = 0
                if x == "":
                    break
            else:
                x = int(x)
                cur += x
    return best
```

```python
max_consec(fn)
```

```python
def top3( fn ):
    with open(fn, "r") as f:
        cals = []
        cur = 0
        while True:
            x = f.readline()
            # test for "" in case no blank line at end of input
            if x == "\n" or x == "":
                print(cur)
                cals.append(cur)
                cur = 0
                if x == "":
                    break
            else:
                x = int(x)
                cur += x

    cals.sort(reverse=True)
    return sum(cals[:3])
```

```python
top3(fn)
```

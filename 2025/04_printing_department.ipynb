# Part 1

```python
class MapaDePapel:
    def __init__( self, text ):
        self.height = len(text)
        self.width = len(text[0])
        self.text = [ [c for c in line] for line in text ]

    def has_paper( self, i, j ):
        if i < 0 or i >= self.height or j < 0 or j >= self.width or self.text[i][j] == ".":
            return False
        else:
            return True
        
    def has_available_paper( self, i, j ):
        if not self.has_paper( i, j ):
            return False
        paper_count = 0
        for n, offset in enumerate( zip( [-1,-1,-1,0,0,1,1,1], [-1,0,1,-1,1,-1,0,1] ) ):
            paper_count += self.has_paper( i+offset[0], j+offset[1] )

            # early stopping
            if paper_count >= 4:
                return False
            if 8-n+paper_count < 4:
                return True
            
        return paper_count < 4
      
    def count_available_paper( self ):
        paper = 0
        for i in range(self.height):
            for j in range(self.width):
                paper += self.has_available_paper( i, j )
        return paper
    
    # Part 2
    def remove_paper( self ):
        tot_count = 0

        while True: # python repeat loop
            iter_count = 0

            for i in range(self.height):
                for j in range(self.width):
                    if self.has_available_paper(i,j):
                        self.text[i][j] = "."
                        iter_count += 1
            tot_count += iter_count
            if iter_count == 0:
                break
        return tot_count

```

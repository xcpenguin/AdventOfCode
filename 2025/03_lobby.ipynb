# Part 1

```python
def max_joltage( line ):
    max1 = max2 = 0
    for i,x in enumerate(line[:-1]):
        if x > max1:
            max1 = x
            max2 = 0
        else:
            max2 = max(max2, x)

    max2 = max(max2, line[-1])

    return max1*10 + max2
```

```python
def part1( puzz ):
    return sum( [max_joltage(line) for line in puzz] )
```

```python
part1(test)
```

```python
part1(puzz)
```

```python
j = 2
tmp[:j] + [10] + [0]*(5-j-1)
```

```python
def max_joltage( line, n=2 ):
    maxes = [0] * n
    L = len(line)
    for i,x in enumerate(line):
        for j,y in enumerate(maxes):
            if i + (n-j) > L:
                continue
            if x > y:
                maxes = maxes[:j] + [x] + [0] * (n-j-1)
                break
    return sum( [x * 10**i for i,x in enumerate(maxes[::-1]) ] )
```

```python
def part2( puzz ):
    return sum( max_joltage(line, 12) for line in puzz )
```

```python
[max_joltage(t, 2) for t in test]
```

```python
[max_joltage(t, 12) for t in test]
```

```python
part2(test)
```

```python
part2(puzz)
```

```python
def parse_text( text ):
    ranges = list()
    ingredients = list()
    for line in text:
        if "-" in line:
            ranges.append( [int(x) for x in line.split("-")] )
        elif len(line) > 0:
            ingredients.append( int(line) )

    return ranges, ingredients
```

```python
text = f'''3-5
10-14
16-20
12-18

1
5
8
11
17
32'''

test = text.split("\n")
test = parse_text( test )
```

```python
fn = "../data/2025/05.txt"
with open(fn, "r") as file:
    text = file.readlines()
puzz = [ line.strip() for line in text ]
puzz = parse_text(puzz)
```

```python
class FoodRange:
    def __init__( self, lo, hi=None ):
        if len(lo) == 2:
            self.values = lo.copy()
        else:
            self.values = [lo, hi]

    def __getitem__( self, i ):
        return self.values[i]
    
    def __lt__( self, other ):
        return (
            (self.values[0] < other.values[0]) or
            (self.values[0] == other.values[0] and self.values[1] == other.values[1])
        )
    
    def __eq__( self, other ):
        return self.values[0] == other.values[0] and self.values[1] == other.values[1]
    
    def __str__( self ):
        return f'''[{str(self.values[0])}, {str(self.values[1])}]'''
    
    def __expr__( self ):
        return self.__str__()
```

```python
def sum_ranges( ranges ):
    tot = 0
    for i, r in enumerate(ranges):
        lo, hi = r[0], r[1]
        for j, s in enumerate(ranges[:i]):
            if lo >= s[0] and lo <= s[1]:
                lo = s[1] + 1
                if lo > hi:
                    break
        if lo <= hi:
            tot += hi - lo + 1
    return tot



def part2( puzz ):
    ranges = [ FoodRange(r) for r in puzz[0] ]
    ranges.sort()
    ans = sum_ranges( ranges )
    return ans

    
```

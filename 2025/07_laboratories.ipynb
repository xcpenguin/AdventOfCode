```python
from collections import defaultdict
```

```python
test_text = f'''.......S.......
...............
.......^.......
...............
......^.^......
...............
.....^.^.^.....
...............
....^.^...^....
...............
...^.^...^.^...
...............
..^...^.....^..
...............
.^.^.^.^.^...^.
...............'''

test = test_text.split("\n")
```

```python
fn = "../data/2025/07.txt"
with open(fn, "r") as file:
    puzz_text = file.readlines()
puzz = [ line[:-1] for line in puzz_text ]
```

# Part 1

```python
def part1( puzz ):
    tachs_above = set()
    splits = 0
    for line in puzz:
        for x, ch in enumerate(line):
            if ch == "S":
                tachs_now = set([x])
            elif ch == "." and x in tachs_above:
                tachs_now.add(x)
            elif ch == "^" and x in tachs_above:
                tachs_now.add(x-1)
                tachs_now.add(x+1)
                splits += 1
        tachs_above = tachs_now
        tachs_now = set()
    return splits
            
```

```python
part1( test )
```

```python
part1( puzz )
```

# Part 2

```python
def part2( puzz ):
    tachs_above = defaultdict(int)
    tach_count = 1
    for line in puzz:
        for x, ch in enumerate(line):
            if ch == "S":
                tachs_now = defaultdict(int)
                tachs_now[x] = 1
            elif ch == "." and x in tachs_above:
                tachs_now[x] += tachs_above[x]
            elif ch == "^" and x in tachs_above:
                tachs_now[x-1] += tachs_above[x]
                tachs_now[x+1] += tachs_above[x]
                tach_count += tachs_above[x]

        tachs_above = tachs_now
        tachs_now = defaultdict(int)
    return tach_count
            
```

```python
part2( test ) 
```

```python
part2( puzz )
```

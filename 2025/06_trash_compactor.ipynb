```python
import re
```

```python
def parse_line( line ):
    return re.sub(r'\s+', ' ', line.strip()).split(" ")

def parse_text( text ):
    values = list()
    n = len( parse_line(text[0]) )
    for i in range(n):
        values.append( list() )

    for line in text:
        parsed = parse_line( line )
        if parsed[0] in ["+", "*"]:
            ops = parsed
        else:
            for i, v in enumerate(parsed):
                values[i].append(int(v))
    return values, ops
```

```python
test_text = f'''123 328  51 64 
 45 64  387 23 
  6 98  215 314
*   +   *   + '''

test_text = test_text.split("\n")
test = parse_text( test_text )
```

```python
fn = "../data/2025/06.txt"
with open(fn, "r") as file:
    puzz_text = file.readlines()
puzz_text = [ line[:-1] for line in puzz_text ] # don't strip whitespace this time. It's important!, just trim the trailing \n
puzz = parse_text( puzz_text )
```

# Part 1

```python
def calc( vals, op ):
    if op == "*":
        tot = 1
        for v in vals:
            tot *= v
    else:
        tot = 0
        for v in vals:
            tot += v
    return tot

def part1( puzz ):
    tot = 0
    for vals, op in zip(puzz[0], puzz[1]):
        tot += calc( vals, op )
    return tot
```

```python
part1(test)
```

```python
part1(puzz)
```

# Part 2

```python
def col_parse( text ):
    # pull the operators out of text
    ops = re.sub( r'\s+', ' ', text[-1]).split(' ')
    val_text = text[:-1]

    values = list()
    new_val = 0

    for col in range(len(val_text[0])):
        # start a new collection of numbers if you found a column of all 0s
        if new_val == 0:
            values.append( list() )
        new_val = 0

        for row in range(len(val_text)):
            if val_text[row][col] != ' ':
                new_val = new_val * 10 + int(val_text[row][col])
        if new_val != 0:
            values[-1].append(new_val)

    return values, ops
        

```

```python
col_parse(test_text)
```

```python
part1( col_parse(test_text) )
```

```python
part1( col_parse(puzz_text) )
```

```python
import re
import numpy as np
```

```python
def parse_text( text ):
    presents = list()
    trees = list()

    for p in range(6):
        presents.append( text[ (p*5)+1 : (p*5)+4 ] )

    for line in text[30:]:
        m = re.fullmatch(r'(\d+)x(\d+):(.*)', line)
        counts = [int(c) for c in m[3].strip().split(" ")]
        trees.append( {"wid":int(m[1]), "len":int(m[2]), "counts":counts } )

    return {"presents":presents, "trees":trees}

    
```

```python
test_text = f'''0:
###
##.
##.

1:
###
##.
.##

2:
.##
###
##.

3:
##.
###
##.

4:
###
#..
###

5:
###
.#.
###

4x4: 0 0 0 0 2 0
12x5: 1 0 1 0 2 2
12x5: 1 0 1 0 3 2'''

test = test_text.split("\n")
test = parse_text(test)
```

```python
fn = "../data/2025/12.txt"
with open(fn, "r") as file:
    puzz_text = file.readlines()
puzz = [ line.strip() for line in puzz_text ]
puzz = parse_text(puzz)
```

# Part 1

```python
def quick_kill( tree, presents ):
    sizes = [ sum( [x=="#" for row in present for x in row] ) for present in presents ]

    needed = sum( c*s for c,s in zip( tree["counts"], sizes) )
    return int( tree["wid"]*tree["len"] < needed )
    return sizes, needed
```

```python
def calc_spare( tree, presents ):
    sizes = [ sum( [x=="#" for row in present for x in row] ) for present in presents ]

    needed = sum( c*s for c,s in zip( tree["counts"], sizes) )
    return tree["wid"]*tree["len"] - needed 
    return sizes, needed
```

```python
bad = [quick_kill(t, puzz["presents"]) for t in puzz["trees"]]
```

```python
trees2 = [ t for i,t in enumerate(puzz["trees"]) if bad[i] == 0 ]
```

```python
len(trees2)
```

```python
spare = [ calc_spare( t, puzz["presents"] ) for t in trees2 ]
np.unique(spare, return_counts=True)
```

It turns out, that the trees are all obviously big enough or obviously too small. Let that be a lesson to look closely at the input!

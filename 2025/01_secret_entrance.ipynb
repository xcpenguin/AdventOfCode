```python
from math import floor
```

```python
def parse_input(text):
    spin = list()
    for line in text:
        x, y = line[0], int(line[1:])
        if x == "L":
            spin.append(-y)
        else:
            spin.append(y)

    return spin
```

```python
text = f'''L68
L30
R48
L5
R60
L55
L1
L99
R14
L82'''

test = parse_input(text.split("\n"))
```

```python
fn = "../data/2025/01.txt"
with open(fn, "r") as file:
    text = file.readlines()
text = [line.strip() for line in text]
puzz = parse_input(text)
```

```python
def partial_sums( arr, start=50 ):
    sums = [start]
    for a in arr:
        sums.append( (sums[-1] + a) % 100 )
    return sums
```

```python
def part1( puzz ):
    count = 0
    for x in partial_sums(puzz):
        count += int(x==0)
    return count
```

# Part 2

```python
# Can we pass 0 multiple times in one turn?  - Yes!
max(puzz), min(puzz)
```

```python
def part2( puzz, start = 50, verbose=False ):
    pos = start
    clicks = 0
    for i, p in enumerate(puzz):
        new_pos = pos + p
        clicks += abs(floor(new_pos/100))
        if verbose:
            print(i, pos, new_pos)

        # edge case 1: start on 0 and go left, we'd get credit for an extra 0
        if pos == 0 and p < 0:
            clicks -= 1
        
        # update pos
        pos = new_pos % 100

        # edge case 2: went left to a zero, we wouldn't get credit for the 0
        if pos == 0 and p < 0:
            clicks += 1
    
        if verbose:
            print(pos, clicks)
    return clicks
```

```python
part2(test, verbose = True)
```
